```python
import pandas as pd

df = pd.read_excel(r"D:\Portfolio Projects\Public Transit Ridership\Data\3_Processed_Data\main_dataset.xlsx", parse_dates=["date"])
df = df.sort_values("date").reset_index(drop=True)

df["ridership_lag1"] = df["ridership"].shift(1)
df["ridership_lag12"] = df["ridership"].shift(12)

df["gas_price_lag1"] = df["gas_price_regular"].shift(1)
df["unemployment_rate_lag1"] = df["unemployment_rate"].shift(1)

df["avg_temp_lag1"] = df["avg_temp"].shift(1)
df["total_precip_mm_lag1"] = df["total_precip_mm"].shift(1)
df["total_snow_cm_lag1"] = df["total_snow_cm"].shift(1)

df_model = df.dropna().copy()

print(df_model.shape)
print(df_model[[
    "date","ridership","ridership_lag1","ridership_lag12",
    "gas_price_lag1","unemployment_rate_lag1"
]].head())
```

```python
df_model.to_csv(r"D:\Portfolio Projects\Public Transit Ridership\Data\3_Processed_Data\main_dataset_modeling.csv", index=False)
df_model.to_excel(r"D:\Portfolio Projects\Public Transit Ridership\Data\3_Processed_Data\main_dataset_modeling.xlsx", index=False)
```
