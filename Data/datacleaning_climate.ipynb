```python
# STEP 1 — Load and append all yearly climate CSV files

from pathlib import Path
import pandas as pd

# folder that contains your 11 climate csv files
data_dir = Path('RAW_Data\climate_daily')  

# get all daily climate csv files
files = sorted(data_dir.glob("en_climate_daily_ON_*_P1D.csv"))

print("Number of files found:", len(files))
for f in files:
    print(f.name)

dfs = []
for file in files:
    df = pd.read_csv(file)
    df["source_file"] = file.name
    dfs.append(df)

climate_raw = pd.concat(dfs, ignore_index=True)

print("\nCombined shape:", climate_raw.shape)
print("\nColumns:")
print(climate_raw.columns.tolist())

display(climate_raw.head())
```

```python
# STEP 2 — First quality checks on combined climate data

print("Duplicate rows:", climate_raw.duplicated().sum())
print("\nMissing values by column:")
print(climate_raw.isna().sum())

# show data types
print("\nData types:")
print(climate_raw.dtypes)

# if Year / Month / Day columns exist, build a proper date
if {"Year", "Month", "Day"}.issubset(climate_raw.columns):
    climate_raw["date"] = pd.to_datetime(
        climate_raw[["Year", "Month", "Day"]]
    )
    print("\nDate range:")
    print(climate_raw["date"].min(), "to", climate_raw["date"].max())

display(climate_raw.head())
```

```python
# STEP 4 — Keep only relevant climate columns and create proper date

climate = climate_raw.copy()

keep_cols = [
    "Date/Time",
    "Year",
    "Month",
    "Day",
    "Mean Temp (°C)",
    "Max Temp (°C)",
    "Min Temp (°C)",
    "Total Rain (mm)",
    "Total Snow (cm)",
    "Total Precip (mm)"
]

climate = climate[keep_cols].copy()

climate["date"] = pd.to_datetime(climate["Date/Time"])
climate = climate.sort_values("date").reset_index(drop=True)

print(climate.shape)
display(climate.head())
```

```python
# STEP 6 — Missing weather values by month

monthly_missing = (
    climate.assign(month_period=climate["date"].dt.to_period("M"))
           .groupby("month_period")[[
               "Mean Temp (°C)",
               "Total Rain (mm)",
               "Total Snow (cm)",
               "Total Precip (mm)"
           ]]
           .apply(lambda x: x.isna().sum())
)

display(monthly_missing.head(15))
print("\nMonths with any missing values:")
display(monthly_missing[(monthly_missing > 0).any(axis=1)])
```

```python
# STEP 7 — Aggregate daily climate to monthly climate features

climate_monthly = (
    climate.assign(date=pd.to_datetime(climate["date"]))
           .groupby(climate["date"].dt.to_period("M"))
           .agg({
               "Mean Temp (°C)": "mean",
               "Max Temp (°C)": "mean",
               "Min Temp (°C)": "mean",
               "Total Rain (mm)": "sum",
               "Total Snow (cm)": "sum",
               "Total Precip (mm)": "sum"
           })
           .reset_index()
)

climate_monthly["date"] = climate_monthly["date"].dt.to_timestamp()

climate_monthly = climate_monthly.rename(columns={
    "Mean Temp (°C)": "avg_temp",
    "Max Temp (°C)": "avg_max_temp",
    "Min Temp (°C)": "avg_min_temp",
    "Total Rain (mm)": "total_rain_mm",
    "Total Snow (cm)": "total_snow_cm",
    "Total Precip (mm)": "total_precip_mm"
})

climate_monthly["year"] = climate_monthly["date"].dt.year
climate_monthly["month"] = climate_monthly["date"].dt.month

print(climate_monthly.shape)
display(climate_monthly.head())
```

```python
# STEP 8 — Monthly completeness check

climate_quality = (
    climate.assign(month_period=climate["date"].dt.to_period("M"))
           .groupby("month_period")
           .agg(
               total_days=("date", "count"),
               missing_temp_days=("Mean Temp (°C)", lambda x: x.isna().sum()),
               missing_precip_days=("Total Precip (mm)", lambda x: x.isna().sum()),
               missing_snow_days=("Total Snow (cm)", lambda x: x.isna().sum())
           )
           .reset_index()
)

climate_quality["date"] = climate_quality["month_period"].dt.to_timestamp()

# number of calendar days in each month
climate_quality["days_in_month"] = climate_quality["date"].dt.days_in_month

# completeness ratios
climate_quality["temp_completeness"] = 1 - climate_quality["missing_temp_days"] / climate_quality["days_in_month"]
climate_quality["precip_completeness"] = 1 - climate_quality["missing_precip_days"] / climate_quality["days_in_month"]
climate_quality["snow_completeness"] = 1 - climate_quality["missing_snow_days"] / climate_quality["days_in_month"]

# simple flag
climate_quality["climate_data_flag"] = (
    (climate_quality["temp_completeness"] < 0.8) |
    (climate_quality["precip_completeness"] < 0.8) |
    (climate_quality["snow_completeness"] < 0.8)
)

display(climate_quality.head(15))
display(climate_quality[climate_quality["climate_data_flag"] == True])
```

```python
# STEP 9 — Merge monthly climate values with climate quality flags

climate_monthly = climate_monthly.merge(
    climate_quality[[
        "date",
        "days_in_month",
        "missing_temp_days",
        "missing_precip_days",
        "missing_snow_days",
        "temp_completeness",
        "precip_completeness",
        "snow_completeness",
        "climate_data_flag"
    ]],
    on="date",
    how="left"
)

display(climate_monthly.head())
```
