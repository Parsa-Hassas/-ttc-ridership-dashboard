```python
import pandas as pd
```

```python
unemp = pd.read_csv(r"D:\Portfolio Projects\Public Transit Ridership\Data\RAW_Data\1410045901_Unemployment(2011-2025)_databaseLoadingData.csv")
unemp.head()
```

```python
unemp.dtypes
```

```python
unemp = unemp.copy()

# convert REF_DATE to proper monthly date
unemp["date"] = pd.to_datetime(unemp["REF_DATE"], format="%Y-%m")

# force month-start timestamp format
unemp["date"] = unemp["date"].dt.to_period("M").dt.to_timestamp()

print(unemp[["REF_DATE", "date"]].head())
print(unemp["date"].dtype)
```

```python
unemp_clean = unemp[['date' , 'GEO' , 'VALUE']]
```

```python
unemp_clean.head()
```

```python
unemp_clean = unemp_clean[unemp_clean['GEO']=='Toronto, Ontario'].copy()
```

```python
unemp_clean = unemp_clean.rename(columns= {'VALUE' : 'Unemployment Rate'})
unemp_clean = unemp_clean.sort_values('date').reset_index(drop=True)
unemp_clean.head()
```

```python
unemp_clean.to_csv(r"D:\Portfolio Projects\Public Transit Ridership\Data\Interim\unemployment_clean.csv", index=False)
```
