```python
import pandas as pd
```

```python
fuel = pd.read_csv(r'D:\Portfolio Projects\Public Transit Ridership\Data\RAW_Data\1810000101_FuelPrice(2007-2025)_databaseLoadingData.csv')
fuel.head()
```

```python
fuel = fuel.copy()

# convert REF_DATE to proper monthly date
fuel["date"] = pd.to_datetime(fuel["REF_DATE"], format="%Y-%m")

# force month-start timestamp format
fuel["date"] = fuel["date"].dt.to_period("M").dt.to_timestamp()

print(fuel[["REF_DATE", "date"]].head())
print(fuel["date"].dtype)
```

```python
fuel_clean = fuel[fuel['GEO']=='Toronto, Ontario']
fuel_clean = fuel_clean[['date' , 'VALUE']]
```

```python
fuel_clean = fuel_clean.rename(columns= {'VALUE' : 'gas_price_regular'})
```

```python
fuel_clean = fuel_clean.sort_values('date').reset_index(drop=True)
```
